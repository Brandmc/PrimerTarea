se tiene las dos desigualdades
$$50x_1+24x_2\leq 2400$$
$$30x_1+33x_2\leq 2100$$

para hacer el plot ls pasamos a igualdades y depsejamos para una varibale

$$50x_1+24x_2= 2400$$
$$30x_1+33x_2= 2100$$

el depsjee queda entonces como:

$$x_2=(2400-50 x_1)/24$$
$$x_2=(2100-30x_1/33$$

$$x_2=\frac{2100-30x_1}{33}$$

```python
def x2_v1(x1):
    return (2400 - 50*x1)/24  # maquina A
def x2_v2(x1):
    return (2100 - 30*x1)/33    # maquina B
```

```python
x1 = np.linspace(0, 100)
```

```python
plt.plot(x1, x2_v1(x1), 'magenta') #desigualdad 1
plt.plot(x1, x2_v2(x1)) # desigualdad 2
plt.plot([45, 45], [0, 25], 'k')
plt.scatter([45,45], [0,25],s=200, color = 'red')
plt.xlim(xmin = 25, xmax = 50)
plt.ylim(ymin = -5, ymax = 45)
plt.plot([43, 47], [5, 5], 'r')
plt.scatter([43,47],[5,5],s=200, color='red')
#plt.scatter([45], [6.25], color = 'red')
#plt.fill_between(np.array([45, 45.6]), x2_v1(np.array([45, 45.6])),
 #               5*np.ones(2), alpha = .2, color = 'orange')
plt.xlabel(r"$x_1$",fontsize=18)
plt.ylabel(r"$x_2$",fontsize=18)
plt.show()
```

## ejercicio monica
Actividad. Mónica hace aretes y cadenitas de joyería. Es tan buena, que todo lo que hace lo vende.

Le toma 30 minutos hacer un par de aretes y una hora hacer una cadenita, y como Mónica también es estudihambre, solo dispone de 10 horas a la semana para hacer las joyas. Por otra parte, el material que compra solo le alcanza para hacer 15 unidades (el par de aretes cuenta como unidad) de joyas por semana.

La utilidad que le deja la venta de las joyas es \$15 en cada par de aretes y \$20 en cada cadenita.

¿Cuántos pares de aretes y cuántas cadenitas debería hacer Mónica para maximizar su utilidad?

$x_1$ son la cantidad de pulseras y $x_2$ son las cantidades de collares. se quiere optimizar $x_1$ y $x_2$

cuantas hora le toma monica hacer aretes y cadenas sin pasar de 10 horas:
$$0.5x_1 + x_2 \leq 10$$
cuantas unidades debe hacer monica para producir  artes y cadenas sin pasar de  15 unidades:
$$x_1 + x_2 \leq 15$$

```python
import numpy as np
import scipy.optimize as opt

#se encuntra el maximo.e pone el valo de la utilidad de cada tipo de joya
c=np.array([-15,-20])

#coeficiente de las variables
A=np.array([[0.5,1],[1,1]])
b=np.array([10,15])
#restriccion de las ecuaciones(solo o pueden ser negativos)
x1_bound=(0,None)
#restriccion de la ecuacion 4
x2_bound=(0,None)
#funciones para solucionar
res=opt.linprog(c,A_ub=A,b_ub=b, bounds=(x1_bound, x2_bound),options={'disp':True})
res
```
